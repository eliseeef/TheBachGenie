```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
# import the personal_info data
contestants = pd.read_csv('./pre_processing/contestants_generalized.csv')
display(contestants.head())
```

## Contestant Hometown

```python
# where all contestants are from

contestant_hometowns = contestants['hometown'].groupby(contestants['hometown']).count()

plt.pie(contestant_hometowns, labels=contestant_hometowns.keys()) 
```

```python
contestant_hometowns.sort_values(ascending=False)
```

```python
# where all winning contestants are from
winning_contestants = contestants[contestants.outcome == 'winner']
contestant_hometowns = winning_contestants['hometown'].groupby(winning_contestants['hometown']).count()

plt.pie(contestant_hometowns, labels=contestant_hometowns.keys()) 
```

```python
contestant_hometowns.sort_values(ascending=False)
```

```python
# where all midExit contestants are from
mid_exit_const = contestants[contestants.outcome == 'midExit']
contestant_hometowns = mid_exit_const['hometown'].groupby(mid_exit_const['hometown']).count()

plt.pie(contestant_hometowns, labels=contestant_hometowns.keys()) 
```

```python
contestant_hometowns.sort_values(ascending=False)
```

```python
# where all earlyExit contestants are from
early_exit_const = contestants[contestants.outcome == 'earlyExit']
contestant_hometowns = early_exit_const['hometown'].groupby(early_exit_const['hometown']).count()

plt.pie(contestant_hometowns, labels=contestant_hometowns.keys()) 
```

```python
contestant_hometowns.sort_values(ascending=False)
```

## Contestant Job

```python
# what occupations all contestants 
contestant_occupation = contestants['job'].groupby(contestants['job']).count()

contestant_occupation_filtered = contestant_occupation[contestant_occupation > 19]

# type(contestant_occupation)

contestant_occupation_filtered.sort_values(ascending=False)
```

```python
plt.pie(contestant_occupation_filtered, labels=contestant_occupation_filtered.keys()) 
```

```python
# what do all winning contestants do
winning_contestants = contestants[contestants.outcome == 'winner']
contestant_occupation = winning_contestants['job'].groupby(winning_contestants['job']).count()

contestant_occupation.sort_values(ascending=False)
```
